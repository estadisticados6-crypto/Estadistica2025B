# Unidad 1. Regresión lineal simple y correlación

## Medidas de dispersión
Suma de los cuadrados de las desviaciones de los valores de $X$ con respecto a su media:
$$ S_{xx} = \sum_{i=1}^{n} (X_i - \bar{X})^2 $$
Suma de los productos de las desviaciones de los valores de $X$ y $Y$ con respecto a sus medias:
$$S_{xy} = \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})$$
Suma de los cuadrados de las desviaciones de los valores de $Y$ con respecto a su media:
$$ S_{xx} = \sum_{i=1}^{n} (Y_i - \bar{Y})^2 $$


## Coeficiente de correlación y de determinación
Coeficiente de correlación de Pearson:
$$ r = \frac{S_{xy}}{\sqrt{S_{xx} \cdot S_{yy}}} $$
Coeficiente de determinación:
$$ r^2 $$

## Recta de regresión ajustada
La regresión lineal ajustada se representa mediante estadísticos:
$$ \hat{Y} = b_0 + b_1 X $$
donde $\hat{Y}$ representa el valor de $Y$ obtenido mediante la recta de regresión ajustada (no la verdadera $Y$). Los estadísticos $b_0$ y $b_1$ se calculan de la siguiente manera:

$$b_1 = \frac{S_{xy}}{S_{xx}}$$

$$ b_0 = \bar{Y} - b_1 \bar{X} $$

## Cálculo de residuales
Residuales:
$$e_i = Y_i - \hat{Y}_i$$

## Sumas de cuadrados SS (Sum of Squares)
Suma de los Cuadrados de los Errores:
$$ SSE = \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2 = \sum_{i=1}^n e_i^2$$
Suma total de cuadrados:
$$SST = \sum_{i=1}^{n} (Y_i - \bar{Y})^2$$
Suma de cuadrados de regresión:
$$SSR = \sum_{i=1}^{n} (\hat{Y_i} - \bar{Y})^2$$
- SST: Mide la variabilidad total de los datos observados.
- SSR: Mide la variabilidad de los datos que el modelo de regresión explica.
- SSE: Mide la variabilidad no explicada por el modelo (es decir, los residuos).


## Intervalo de confianza
Estadístico de prueba _t_:
$$t = \frac{b_1}{SE(b_1)}$$
Error estándar de $b_1$:
$$SE(b_1) = \frac{\sqrt{SSE / (n-2)}}   {\sqrt{S_{xx}} }$$
o
$$SE(b_1) = \frac{\sqrt{\sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2  / (n-2)}}   {\sqrt{S_{xx}} }$$

Intervalo de confianza para $b_1$:
$$b_1 - t_{\alpha/2} \cdot SE(b_1) \lt \beta_1 \lt b_1 + t_{\alpha/2} \cdot SE(b_1)$$
donde $n$ representa la cantidad de pares de datos.

## Comprobación de supuestos
Comprobar suposiciones:
- Test de shapiro a los residuales $e_i$: Para comprobar si la distribución es normal sobre la recta.
- Grafico $X$ vs $Y$: Para observar si los datos soportan la suposición de linealidad.
- Gráfico de residuales: Para observar si los datos soportan la suposición de linealidad, complementario al coeficiente de correlación
- Test de Breusch-Pagan: Para detectar heteroscedasticidad en regresion lineal

Test de Shapiro:<br>
`from scipy.stats import shapiro`<br>
Después, se obtiene el valor-p:<br>
`_, valor_p_sh = shapiro(data)`<br>
- $H_0$: Los datos siguen una distribución normal
- $H_1$: Los datos no siguen una distribución normal

Test de Breusch-Pagan:<br>
`from statsmodels.stats.api import het_breuschpagan`<br>
Después, se obtiene el valor-p:<br>
`_, valor_p_bp, _, _ = het_breuschpagan(residuales, X)`
- $H_0$: Hay homoscedasticidad
- $H_1$: Hay heteroscedasticidad

## ANOVA en regresión lineal

| Fuente de variación | Suma de cuadrados (SS) | Grados de libertad (df) | Promedio de los cuadrados (MS) | Estadístico F |
|---------------------|------------------------|--------------------------|--------------------------------|---------------|
| Regresión  | $ SSR$ | $p$ | $MSR = \frac{SSR}{p}$ | $F = \frac{MSR}{MSE}$|
| Error | $SSE$| $n - p - 1 $| $MSE = \frac{SSE}{n - p - 1} $|     |
| Total | $SST $| $n - 1$|             |        |

donde $p$ es el número de parámetros para la recta de regresión ajustada (en la regresión simple $p=1$). Las hipótesis son:
$$H_0: \beta_1 = 0$$
$$H_1: \beta_1 \ne 0 $$

```python
import pandas as pd
df=pd.DataFrame({
    "examenes_cortos": [59, 92, 72, 90, 95, 87, 89, 77, 76, 65, 97, 42, 94, 62, 91],
    "promedio_final": [64, 84, 77, 80, 77, 81, 80, 84, 80, 69, 83, 40, 78, 65, 90]
})
#a) declarar variable dependiente e independiente

x=df["examenes_cortos"]
y=df["promedio_final"]



#b)diagrama de dispersión
import matplotlib.pyplot as plt
#figsize es la escala del cuadro
#dpi es la resolución del cuadro
plt.figure(figsize=(10,7),dpi=100)
plt.scatter(x , y, label="calificación", color="blue", s=30)
plt.xlabel("examenes_cortos", fontsize=20)
plt.ylabel("promedio_final", fontsize=20)
#plt.xlabel es para ponerle un titulo en la parte de la x o ylabel en la y
plt.title("grafico de dispersión",  fontsize=30, color="red")
##plt.title es para colocarle un titulo a toda la grafica
plt.gca().spines[["top","right"]]. set_visible(False)
##lo anterior es para quitar las lineas de arriba y la derecha para que se vea solo lo de x y y
plt.grid()
##plt.grid para cuadricular
plt.legend()
##plt.legend para colocar la leyenda de los puntos


##c) soporta la suposicion de linealidad
print("los datos soportan la suposición de linealidad")


##d) Calcule el coeficiente de correlación e interprete el resultado
# el coefiiente de correlacion parametro es P y estadistico r
# el coeficientcoeficiente de correlacion suele ser : -1<=r=<1
Sxx=((x-x.mean())**2).sum()
print("Sxx=",Sxx)
Syy=((y-y.mean())**2).sum()
print("Syy=",Syy)
Sxy=((x-x.mean())*(y-y.mean())).sum()
print("Sxy=",Sxy)
from math import sqrt
r=Sxy/sqrt(Sxx*Syy)
print("coeficiente de correlacion=",round(r,4))
##round(r, 4) sirve para limitar la cantidad de decimales que se requiere
#print("coeficiente de correlacion (modo facil)",df.corr())
from scipy.stats import pearsonr
print("coeficiente de correlacion (scipy)",pearsonr(x,y))
#SE USARA ESTA PARA SACAR R
##e) Calcule el coeficiente de determinación e interprete el resultado
r2=r**2
print("coeficiente de determinación=",r2)
print("EL 75% de los examenes cortos predicen a la calificacion de los examenes finales")
##hay otra forma para sacar el coeficiente de determinacion,
# from sklearn.sklearn.metries r2r2_score
# r2=r2r2_score(y,yyc)
##PERO ES PARA DESPUES
##f) Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión
import statsmodels.api as sm
x_constante=sm.add_constant(x)
modelo=sm.OLS(y,x_constante).fit()
modelo.params
print("parametros:",modelo.params)
yc=modelo.predict(x_constante)
plt.plot(x,yc, color="black", linestyle="--", linewidth=2, alpha=0.5)
plt.show()
##quitamos el plt.show de arriba para juntar las dos tablas, figure es el comienzo y hasta que no se coloque el show todo sera parte de la grafgrafica
from sklearn.metrics import r2_score
r2=r2_score(y,yc)
print("coeficiente de determinación= ",round(r2,4))

print("intervalo de confianza")
##g) Obtenga un intervalo de confianza del 95% para la pendiente de la recta de regresión ajustada ($b_1$)
#obtener el pecesito
intervalo_de_confianza=modelo.conf_int(alpha=0.05)
intervalo_de_confianza
print("\n intervalo de confianza:",intervalo_de_confianza)
#\n sirve para altar espacio
#8
residuales=modelo.resid
plt.figure(figsize=(10,7),dpi=100)
plt.scatter(x , y,label="residuales", color="blue", s=30)
plt.xlabel("examenes cortos", fontsize=20)
plt.ylabel("residuales", fontsize=20)
plt.title("grafico de residuales",  fontsize=30, color="red")
plt.show()
plt.axhline(y=0, color="red", linestyle="--")
#test de shapiro es para normalidad Ho= hay normalidad H1=no hay normalidad pez=0.05
#test breusch-pagan homoscedasticidad Ho= hay homoscedasticidad H1= no hay homoscedasticidad pez=0.05
from scipy.stats import shapiro
#9
estadisitico_shapiro,valor_p_sh = shapiro(residuales)
print("valor p (shapiro) =",valor_p_sh)
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, x_constante)
print(f'valor_p de Breusch-Pagan: {valor_p_bp:0.4f}\n')
modelo.params
#f(x)=24.52 + 0.64x
#f(70)=24.52+0.64 (70)
print("x=70,", modelo.predict([1,70]))
from statsmodels.formula.api import ols
modelo_lineal = ols("promedio_final ~ examenes_cortos", data=df).fit()
tabla_anova=sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

5. Calcule el coeficiente de determinación e interprete el resultado.
6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
7. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de regresión ajustada ($b_1$)

```python
import pandas as pd
df=pd.DataFrame({
    "categoria_salario": [11, 10, 8, 5, 9, 7, 3, 11, 8, 7, 2, 9, 8, 3],
    "categoria_ausencia": [18, 17, 29, 36, 11, 28, 35, 14, 20, 32, 39, 16, 31, 40]
})

##1) variable independiente y dependiente
#siempre y= dependiente x= independiente
# ~=depende de (salario~ausencia)
x=df["categoria_salario"]
y=df["categoria_ausencia"]

##2) diagrama de dispercion
import matplotlib.pyplot as plt
#generar graficos, visualizaciones esteticas
plt.figure(figsize=(10,10),dpi=100)
#figsize es la escala del cuadro
#dpi es la resolución del cuadro
plt.scatter(x , y, label="correlacion", color="blue", s=20)
#plt.scatter= visualizar la relacion entre dos variables
plt.xlabel("categoria_salario", fontsize=15)
plt.ylabel("categoria_ausencia", fontsize=15)
#plt.xlabel es para ponerle un titulo en la parte de la x o ylabel en la y
plt.title("grafico de dispersión",  fontsize=20, color="red")
#plt.title es para colocarle un titulo a toda la grafica
plt.gca().spines[["top","right"]]. set_visible(False)
#lo anterior es para quitar las lineas de arriba y la derecha para que se vea solo lo de x y y
plt.grid()
#plt.grid para cuadricular
plt.legend()
#plt.legend para colocar la leyenda de los puntos

##3)supocision de linealidad
print("\n Los datos soportan la suposición de linealidad")

##4)coeficiente de correlación
# el coefiiente de correlacion= parametro es P y estadistico r
# el coeficientcoeficiente de correlacion suele ser : -1<=r=<1
Sxx=((x-x.mean())**2).sum()
print("\n Sxx=",Sxx)
Syy=((y-y.mean())**2).sum()
print("\n Syy=",Syy)
Sxy=((x-x.mean())*(y-y.mean())).sum()
print("\n Sxy=",Sxy)
from math import sqrt
r=Sxy/sqrt(Sxx*Syy)
print("\n Coeficiente de correlacion=",round(r,4))
##round(r, 4) sirve para limitar la cantidad de decimales que se requiere a 4
#print("coeficiente de correlacion (modo facil)",df.corr())
from scipy.stats import pearsonr
print("\n Coeficiente de correlacion (scipy)",pearsonr(x,y))
#tambien se puede usar para sacar r

##5)coeficiente de determinación
#
r2=r**2
print("\n Coeficiente de determinación=",r2)
print("\n El 77.46% de salarios es por ausencia")
#hay otra forma para sacar el coeficiente de determinacion,
#from sklearn.sklearn.metries r2r2_score
#r2=r2r2_score(y,yyc)

##6)recta de regresión ajustada y reprecentacion sobre el gráfico de dispersión.
import statsmodels.api as sm
x_constante=sm.add_constant(x)
modelo=sm.OLS(y,x_constante).fit()
modelo.params
print("\n parametros:",modelo.params)
yc=modelo.predict(x_constante)
plt.plot(x,yc, color="black", linestyle="--", linewidth=2, alpha=0.5)
plt.show()
##quitamos el plt.show de arriba para juntar las dos tablas, figure es el comienzo y hasta que no se coloque el show todo sera parte de la grafgrafica
from sklearn.metrics import r2_score
r2=r2_score(y,yc)
print("\n coeficiente de determinación= ",round(r2,4))

##7)intervalo de confianza del 95% para la pendiente de la recta de regresión ajustada ( b1 )
intervalo_de_confianza=modelo.conf_int(alpha=0.05)
intervalo_de_confianza
print("\n intervalo de confianza:",intervalo_de_confianza)
#\n sirve para altar espacio

##8)calcule los residuales y trace un nuevo gráfico de dispersión
residuales=modelo.resid
plt.figure(figsize=(10,10),dpi=100)
plt.scatter(x , y,label="residuales", color="blue", s=20)
plt.xlabel("categoria_salario", fontsize=20)
plt.ylabel("categoria_ausencias", fontsize=20)
plt.title("grafico de residuales",  fontsize=20, color="red")
plt.show()
plt.axhline(y=0, color="red", linestyle="--")
#test de shapiro es para normalidad Ho= hay normalidad H1=no hay normalidad pez=0.05
#test breusch-pagan homoscedasticidad Ho= hay homoscedasticidad H1= no hay homoscedasticidad pez=0.05

##9)prueba de Shapiro para los residuales
estadisitico_shapiro,valor_p_sh = shapiro(residuales)
print("valor p (shapiro) =",valor_p_sh)

##10)prueba de Brausch-Pagan para los residuales
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, x_constante)
print(f'valor_p de Breusch-Pagan: {valor_p_bp:0.4f}\n')
```

```python
import pandas as pd
df=pd.DataFrame({
    "construccion_casas": [2.0, 2.5, 3.2, 3.6, 3.7, 4.0, 4.2, 4.6, 4.8, 5.0],
    "linea_blanca": [5.0, 5.5, 6.0, 7.0, 7.2, 7.7, 8.4, 9.0, 9.7, 10.0]
})

##1) variable independiente y dependiente
#siempre y= dependiente x= independiente
# ~=depende de (salario~ausencia)
x=df["construccion_casas"]
y=df["linea_blanca"]

##2) diagrama de dispercion
import matplotlib.pyplot as plt
#generar graficos, visualizaciones esteticas
plt.figure(figsize=(10,10),dpi=100)
#figsize es la escala del cuadro
#dpi es la resolución del ntsize=15)
#plt.xlabel es para ponerle un titulo en la parte de la x o ylabel en la y
plt.title("grafico de dispersión",  fontsize=20, color="red")
#plt.title es para colocarle un titulo a toda la grafica
plt.gca().spines[["top","right"]]. set_visible(False)
#lo anterior es para quitar las lineas de arriba y la derecha para que se vea solo lo de x y y
plt.grid()
#plt.grid para cuadricular
plt.legend()
#plt.legend para colocar la leyenda de los puntos

##3)supocision de linealidad
print("\n Los datos soportan la suposición de linealidad")

##4)coeficiente de correlación
#print("coeficiente de correlacion (modo facil)",df.corr())
from scipy.stats import pearsonr
print("\n Coeficiente de correlacion (scipy)",pearsonr(x,y))
#tambien se puede usar para sacar r

##6)recta de regresión ajustada y reprecentacion sobre el gráfico de dispersión.
#primero hay que hacer el 6 para que salga el 5
import statsmodels.api as sm
x_constante=sm.add_constant(x)
modelo=sm.OLS(y,x_constante).fit()
modelo.params
print("\n parametros:",modelo.params)
yc=modelo.predict(x_constante)
plt.plot(x,yc, color="black", linestyle="--", linewidth=2, alpha=0.5)
plt.show()
##quitamos el plt.show de arriba para juntar las dos tablas, figure es el comienzo y hasta que no se coloque el show todo sera parte de la grafgrafica

##5)coeficiente de determinación
#hay otra forma para sacar el coeficiente de determinacion
from sklearn.metrics import r2_score
r2=r2_score(y,yc)
print("\n coeficiente de determinación= ",round(r2,4))

##7)intervalo de confianza del 95% para la pendiente de la recta de regresión ajustada ( b1 )
intervalo_de_confianza=modelo.conf_int(alpha=0.05)
print("\n intervalo de confianza:",intervalo_de_confianza)
print("\n 1.46 < beta > 2.02)")
print("\n como es diferente de cero y positiva, SI HAY correlacion lineal")
# los datos salen de la tabla de intervalo de confianza, variable independiente
#\n sirve para altar espacio

##8)calcule los residuales y trace un nuevo gráfico de dispersión
residuales=modelo.resid
plt.figure(figsize=(10,10),dpi=100)
plt.scatter(x , y,label="residuales", color="blue", s=20)
plt.xlabel("construccion_casas", fontsize=20)
plt.ylabel("linea_blanca", fontsize=20)
plt.title("grafico de residuales",  fontsize=20, color="red")
plt.show()
plt.axhline(y=0, color="red", linestyle="--")
#test de shapiro es para normalidad Ho= hay normalidad H1=no hay normalidad pez=0.05
#test breusch-pagan homoscedasticidad Ho= hay homoscedasticidad H1= no hay homoscedasticidad pez=0.05

##9)prueba de Shapiro para los residuales
estadisitico_shapiro,valor_p_sh = shapiro(residuales)
print("valor p (shapiro) =",valor_p_sh)

##10)prueba de Brausch-Pagan para los residuales
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, x_constante)
print(f'valor_p de Breusch-Pagan: {valor_p_bp:0.4f}\n')
```
