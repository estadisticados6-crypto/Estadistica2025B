```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/cesarinrc07-ship-it/Estad-stica2025B/refs/heads/main/healthcare-dataset-stroke-data.csv")
x=df["age"]
y=df["stroke"]

import matplotlib.pyplot as plt
plt.figure(figsize=(15,9), dpi=100)
plt.scatter(x , y, label="Salary", color="blue", s=30)
plt.xlabel("age", fontsize=20)
plt.ylabel("stroke", fontsize=20)
plt.title("grafico de dispersión",  fontsize=30, color="red")
plt.gca().spines[["top","right"]].set_visible(False)
plt.grid()


print("los datos soportan la suposición de linealidad")

#print("coeficiente de correlacion (modo facil)",df.corr())
from scipy.stats import pearsonr
print("\n coeficiente de correlacion (scipy)",pearsonr(x,y))


import statsmodels.api as sm
x_constante=sm.add_constant(x)
modelo=sm.OLS(y,x_constante).fit()
modelo.params
print("parametros:",modelo.params)
yc=modelo.predict(x_constante)
plt.plot(x,yc, color="black", linestyle="--", linewidth=2, alpha=0.5)
plt.show()

##coeficiente de determinacion
from sklearn.metrics import r2_score
r2=r2_score(y,yc)
print("\n coeficiente de determinación=",round(r2,4))

##intervalo de confianza
intervalo_de_confianza= modelo.conf_int(alpha=0.05)
print("\n -ntervalo de confianza",intervalo_de_confianza)
print("8674.11< beta < 10225.8")
print("Como es diferente a 0, entonces si hay correlacion lineal")
#residuales
residuales=modelo.resid
plt.figure(figsize=(15,9),dpi=100)
plt.scatter( x ,residuales, label="residuales", color="blue", s=30)
plt.xlabel("age", fontsize=20)
plt.ylabel("stroke", fontsize=20)
plt.title("Residuales",  fontsize=30, color="red")
plt.gca().spines[["top","right"]]. set_visible(False)
plt.grid()
plt.axhline(y=0, color="grey", linestyle="--")
plt.show()


from scipy.stats import shapiro
estadistico_shapiro, valor_p_shapiro =shapiro(residuales)
print("\n valor-p (shapiro)",valor_p_shapiro)
print("Se aprueba la H0: hay normalidad")

from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, x_constante)
print(f'valor_p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')
print("x=2,",modelo.predict([1,2]))
print("el 43748 es el correspondiente al salario con 2 años de experiencia")
print("x=1.5,",modelo.predict([1,1.5]))
print("el 7 es el correspondiente al salario con 1.5 años de experiencia")

from statsmodels.formula.api import ols
mmodelo_lineal= ols("stroke~age", data=df).fit()
table_anova=sm.stats.anova_lm(mmodelo_lineal)
print(table_anova)
print("la ultima fila es el valo-p")
print("entonces se rechaza la H0 y H1 diferente 0 se aprueba")
```
